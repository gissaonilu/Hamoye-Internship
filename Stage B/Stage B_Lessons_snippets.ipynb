# Stage B: Machine Learning_Regression

## Import Libraries

```python
# import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

## Import Datasets

```python
#read the datasets for the Machine Learning Regression
df = pd.read_excel('https://archive.ics.uci.edu/ml/machine-learning-databases/00242/ENB2012_data.xlsx')
df.head(10)
```

## Rename columns

```python
#rename   columns
column_names   =   {'X1':'Relative_Compactness','X2':'Surface_Area','X3':'Wall_Area',
                    'X4':'Roof_Area','X5':'Overall_Height','X6':'Orientation',  
                    'X7':'Glazing_Area','X8':'Glazing_Area_Distribution',
                    'Y1':'Heating_Load','Y2':'Cooling_Load' }
df.head()
```

```python
df   =   df.rename(columns=column_names)
#select   a   sample   of   the   dataset
simple_linear_reg_df   =   df[['Relative_Compactness' , 'Cooling_Load']].sample(15, random_state=2)
simple_linear_reg_df
```

## Plot a linear regression graph

```python
#regression   plot
sns.regplot(x="Relative_Compactness",y="Cooling_Load",data=simple_linear_reg_df) 
```
